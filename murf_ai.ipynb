**Quickstart**

```python
from murf import Murf
client = Murf(
    api_key="YOUR_KEY"
)
res = client.text_to_speech.generate(
    text="There is much to be said",
    voice_id="en-US-terrell",
)
print(res.audio_file)
```

**Adding Styles**

```python
from murf import Murf
client = Murf(
    api_key="YOUR_KEY"
)
res = client.text_to_speech.generate(
    text="Oh! I'll have to do this all over again.",
    voice_id="en-US-ken",
    style="Angry"
)
```

**Custom Pronunciations**

```python
from murf import Murf
client = Murf(
    api_key="YOUR_KEY"
)
res = client.text_to_speech.generate(
    text="The 2010 world cup was held in South Africa",
    voice_id="en-US-natalie",
    pronunciation_dictionary={
        "live": { "type": "IPA", "pronunciation": "laÉªv" },
        "2010": { "type": "SAY_AS", "pronunciation": "two thousand and ten" }
    }
)
```

**MultiNative Voices**

```python
from murf import Murf
client = Murf(
    api_key="YOUR_KEY"
)
res = client.text_to_speech.generate(
    text="Croissant",
    voice_id="en-US-natalie",
    multi_native_locale="fr-FR"
)
```

**Custom Audio Duration**

```python
from murf import Murf
client = Murf(
    api_key="YOUR_KEY"
)
res = client.text_to_speech.generate(
    text="The team is down by three points. Ten seconds left on the clock! The next play could decide the game",
    voice_id="en-US-miles",
    audio_duration=8.0
)
```
